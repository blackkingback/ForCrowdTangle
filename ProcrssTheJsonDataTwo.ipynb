```python
import pandas as pd
import json
from datetime import datetime
from dateutil import tz
```

```python
f = open("posts.json","r")
posts_json = json.loads(json.load(f))
```

```python
extracted_info_list = []

for ele in posts_json:
    temp_dict = {}

    FB_account = ele['account']['name']
    temp_dict['FB_account'] = FB_account

    FB_ID = ele['account']['id']
    temp_dict['FB_ID'] = FB_ID

    FB_platformID = ele['account']['platformId']
    temp_dict['FB_platformID'] = FB_platformID

    try:
        FB_handle = ele['account']['handle']
    except KeyError as e:
        FB_handle = None
    temp_dict['FB_handle'] = FB_handle

    PageCategory = ele['account']['pageCategory']
    temp_dict['PageCategory'] = PageCategory

    try:
        PageAdminTopCountry = ele['account']['pageAdminTopCountry']
    except KeyError as e:
        PageAdminTopCountry = None
    temp_dict['PageAdminTopCountry'] = PageAdminTopCountry

    try:
        PageDescription = ele['account']['pageDescription']
    except KeyError as e:
        PageDescription = None
    temp_dict['PageDescription'] = PageDescription

    try:
        PageCreated = ele['account']['pageCreatedDate']
    except KeyError as e:
        PageCreated = None
    temp_dict['PageCreated'] = PageCreated

    PostCreated = ele['date']
    temp_dict['PostCreated'] = PostCreated

    PostCreatedDate = datetime.strptime(PostCreated, '%Y-%m-%d %H:%M:%S').date()
    temp_dict['PostCreatedDate'] = PostCreatedDate

    PostCreatedTime = datetime.strptime(PostCreated, '%Y-%m-%d %H:%M:%S').time()
    temp_dict['PostCreatedTime'] = PostCreatedTime


    # METHOD 1: Hardcode zones:
    from_zone = tz.gettz('UTC')
    to_zone = tz.gettz('tw')
    utc_created_time = datetime.strptime(PostCreated, '%Y-%m-%d %H:%M:%S')
    utc_created_time = utc_created_time.replace(tzinfo=from_zone)
    converted_time = utc_created_time.astimezone(to_zone)
    PostCreatedDate_tw = converted_time.date()
    PostCreatedTime_tw = converted_time.time()
    temp_dict['PostCreated_tw'] = converted_time
    temp_dict['PostCreatedDate_tw'] = PostCreatedDate_tw
    temp_dict['PostCreatedTime_tw'] = PostCreatedTime_tw

    Type = ele['type']
    temp_dict['Type'] = Type

    Likes = ele['statistics']['actual']['likeCount']
    temp_dict['Likes'] = Likes

    Comments = ele['statistics']['actual']['commentCount']
    temp_dict['Comments'] = Comments

    Share = ele['statistics']['actual']['shareCount']
    temp_dict['Share'] = Share

    Love = ele['statistics']['actual']['loveCount']
    temp_dict['Love'] = Love

    Wow = ele['statistics']['actual']['wowCount']
    temp_dict['Wow'] = Wow

    Haha = ele['statistics']['actual']['hahaCount']
    temp_dict['Haha'] = Haha

    Sad = ele['statistics']['actual']['sadCount']
    temp_dict['Sad'] = Sad

    Angry = ele['statistics']['actual']['angryCount']
    temp_dict['Angry'] = Angry

    Care = ele['statistics']['actual']['careCount']
    temp_dict['Care'] = Care

    TotalInteractions = Likes + Comments + Share + Love + Wow + Haha + Sad + Angry + Care
    temp_dict['TotalInteractions'] = TotalInteractions

    URL = ele['postUrl']
    temp_dict['URL'] = URL

    try:
        Message = ele['message']
    except KeyError as e:
        Message = None
    temp_dict['Message'] = Message

    try:
        Link = ele['link']
    except KeyError as e:
        Link = None
    temp_dict['Link'] = Link

    try:
        FinalLink = ele['expandedLinks'][0]['expanded']
    except KeyError as e:
        FinalLink = None
    temp_dict['FinalLink'] = FinalLink

    OverPerformingScore = TotalInteractions
    temp_dict['OverPerformingScore'] = TotalInteractions

    try:
        Description = ele['description']
    except KeyError as e:
        Description = None
    temp_dict['Description'] = Description

    extracted_info_list.append(temp_dict)
```

```python
df = pd.DataFrame(extracted_info_list)
```
