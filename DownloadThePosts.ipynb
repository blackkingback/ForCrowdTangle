```python
import datetime
import time

begin = datetime.date(2020,1,1)
end = datetime.date(2023,6,20)
delta = datetime.timedelta(2)
d = begin

results_list = []

while d<=end:
    temp_start = d
    temp_end = d + delta

    data = pct.ct_get_posts(api_token=api_token,count=100,start_date=temp_start,end_date=temp_end)
    temp_target = data['result']['posts']
    results_list = results_list + temp_target
    print(temp_start,temp_end,data['status'],len(temp_target),len(results_list))
    time.sleep(12)
    d = d + delta
```

```python
import json

final = json.dumps(results_list,indent=2)
```

```python
with open("posts.json", "w") as target:
   json.dump(final, target)
```
