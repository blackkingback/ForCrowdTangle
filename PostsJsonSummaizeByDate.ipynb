```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("ResultsWithFormat.csv",sep="\t")
df['PostCreated'] = pd.to_datetime(df['PostCreated'])
```

```python
df.head(5)
```

```python
results_list = []
for name,temp_group in df.groupby([df['FB_account'], df['PostCreated'].dt.year.rename('year'), df['PostCreated'].dt.month.rename('month')]):
    temp_dict = {}
    temp_dict["Account name"] = name[0]
    temp_dict["Year"] = name[1]
    temp_dict["Month"] = name[2]
    temp_dict["Post_Count"] = len(temp_group)
    results_list.append(temp_dict)
results_df = pd.DataFrame(results_list)
results_df.to_csv("PostsSummarizeByMonth.csv",index=False,sep="\t")
```
